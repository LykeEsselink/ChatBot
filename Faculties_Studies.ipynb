```python
import urllib.request as urllib
from bs4 import BeautifulSoup
import numpy as np
import json
```

```python
with open('uva_json_file', 'r') as f:
    data = json.load(f)
```

```python
faculty_studies = {}
for faculty, studies in data.items():
    studies_list = []
    for study, az_links in studies.items():
        studies_list.append(study)
    faculty_studies[faculty] = studies_list
```

```python
f = open("faculty_studies.txt","w")
f.write( str(faculty_studies) )
f.close()
```
